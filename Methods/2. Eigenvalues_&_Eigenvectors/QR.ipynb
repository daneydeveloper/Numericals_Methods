# Operações Básica

```
def matrixByMatrix(a,b):
    na = len(a)
    ma = len(a[0])
    nb = len(b)
    mb = len(b[0])
    
    mid = min(ma, nb)
    
    if ma != nb:
        raise Exception('Matrix Error!')
    
    c = [[0 for j in range(mb)] for i in range(na)]
    
    for i in range(na):
        for j in range(mb):
            for k in range(mid):
                c[i][j] += a[i][k] * b[k][j]
    
    return c
```

```
def matrixLHNorm2(a):
    n = len(a)
    m = len(a[0])
    soma = 0.0
    for j in range(m-1):
        for i in range(j+1,n):
            soma += a[i][j] * a[i][j]
    return soma
```

```
def Identidade(n,m):
    I = [[0 for i in range(m)] for j in range(n)]
    for i in range(min(n,m)):
        I[i][i] = 1
    return I
```

```
def Transposta(A):
    n = len(A)
    m = len(A[0])
    
    B = [[0 for j in range(n)] for i in range(m)]
    for i in range(n):
        for j in range(m):
            B[j][i] = A[i][j]
    return B
```

```
def printMatrix(A):
    for i in A:
        for j in i:
            s = "%.5f" % j
            print("%8s" % s, end = " ")
        print()
```

# Método de QR

```
def decomposicaoQR(A):
    tol = 0.000000000001

    n = len(A)
    m = len(A[0])
    Q = Identidade(m,n)
    R = [[i for i in j] for j in A]
    
    for j in range(0,n-1):
        for i in range(j+1,n):
            if(abs(R[i][j]) >= tol):
                
                if(abs(R[j][j]) < tol):
                    ang = PI/2
                else:
                    ang = atan(R[i][j]/R[j][j])

                P = Identidade(n,m)
                P[i][i] =  cos(ang)
                P[j][j] =  cos(ang)
                P[i][j] =  sin(ang)
                P[j][i] = -sin(ang)

                Pt = Transposta(P)

                Q = matrixByMatrix(Q,P)
                R = matrixByMatrix(Pt,R)
    
    return [Q,R]
```

```
def metodoQR(A, e):
    n = len(A)
    m = len(A[0])
    
    P = Identidade(n,m)
    A_new = [[i for i in j] for j in A]
    var_norm = e + 1
    
    while var_norm >= e:
        Q, R = decomposicaoQR(A_new)
        A_new = matrixByMatrix(R,Q)
        P = matrixByMatrix(P,Q)
        
        var_norm = matrixLHNorm2(A_new)
    
    return [A_new, P]
```

# Matriz e Tolerância

```
A = [
      [37,  4,  3,  3,  3],
      [ 4, 21,  8,  3,  7],
      [ 3,  8, 56,  6,  6],
      [ 3,  3,  6, 36, 11],
      [ 3,  7,  6, 11, 57]
]

tolerance = 0.000000001
```

# Resultados

```
auto_valor, auto_vetor = metodoQR(A, tolerance)
print("Auto valores:")
printMatrix(auto_valor)
print("Auto vetores:")
printMatrix(auto_vetor)
```
